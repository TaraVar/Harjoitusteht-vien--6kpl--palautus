```python
# first lets create a function to define the price ranges
def categorize_price(rating):
    if rating < 5:
        return 'Low'
    elif rating < 10:
        return 'Medium'
    else:
        return 'High'
# We can use the .apply() method to create a new column 'price_group' based on the 'item_price' column.
threads_reviews['source'] = threads_reviews['rating'].apply(categorize_price)
threads_reviews.head()
```
