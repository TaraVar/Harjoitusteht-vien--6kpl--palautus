```python
import pandas as pd
```

```python
ratings_threads_reviews1 = {"rating" : [2, 1, 3], "review_date" : ["2023-07-08 14:18:24", "2023-07-19 19:35:31","2023-07-10 00:53:25"],"source" :["Google Play", "Google Play", "Google Play"]}
reviews_df = pd.DataFrame(ratings_threads_reviews1)
```

```python
ratings_threads_reviews2 = {"rating" : [4, 2, 5], "review_date" : ["2023-07-07 20:00:49", "2023-07-09 01:06:39", "2023-07-06 18:39:12"], "source" : ["Google Play", "Google Play", "Google Play"]}
ratings_df = pd.DataFrame(ratings_threads_reviews2)
```

```python
combined_df = pd.concat([reviews_df, ratings_df], ignore_index=True)


print(combined_df)
```
