```python
import random
thrownDiceNumbers = []

for i in range(5):
    thrownDiceNumbers.append(random.randint(1, 6))
```

```python
print(thrownDiceNumbers)
print(sum(thrownDiceNumbers))
print(max(thrownDiceNumbers))
```
